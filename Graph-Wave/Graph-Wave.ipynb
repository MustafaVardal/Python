```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
x = np.linspace(-1*np.pi, 1*np.pi, 256, endpoint=True)
```

```python
sinx= np.sin(x)
cosx= np.cos(x)
```

```python
plot.plot(x, sinx)
plot.plot(x, cosx)
plot.title('Sinx & Cosx wave')
plot.xlabel('Time')
plot.ylabel('Width')
plot.grid(True, which= 'both')
plot.axhline(y=0, color='k')
plot.show()
```
