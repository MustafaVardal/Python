```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import gridspec
```

```python
data = pd.read_csv("creditcard.csv")
```

```python
data.head()
```

```python
print(data.shape)
print(data.describe())
```

```python
fraud = data[data['Class'] == 1]
valid = data[data['Class'] == 0]
outlierFraction = len(fraud)/float(len(valid))
print(outlierFraction)
print('Fraud Cases: {}'.format(len(data[data['Class']== 1 ])))
print('Valid Transaction: {}'.format(len(data[data['Class']==0])))
```

```python
print("Details of the fraud transactions: ")
fraud.Amount.describe()
```

```python
print("Details of valid transaction: ")  #Valid transaction calculate the prediction...
valid.Amount.describe()
```

```python
#Using the correlation matrix gives is it correlate to other features or no...
#INSTANCE: V10 AND V8 are negatively feature in Amount...
corrdata= data.corr()
fig = plt.figure(figsize=(12,9))
sns.heatmap(corrdata, vmax=.8,square = True)
plt.show()
```
